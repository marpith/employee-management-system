```python
import pymongo
import bson
from bson import ObjectId
```

```python
client = pymongo.MongoClient('mongodb://localhost:27017/')

db = client['employee_management_system']


collection = db['employees']

print("\t\t\t\t REVA Employee Management System")

def main_menu():
    print('\nEmployee Management System')
    print('===========================')
    print('1. Add an employee')
    print('2. List all employees')
    print('3. View employee details')
    print('4. Update an employee')
    print('5. Delete an employee')
    print('0. Exit')
    choice = input('Enter your choice: ')

    if choice == '1':
        add_employee()
    elif choice == '2':
        list_employees()
    elif choice == '3':
        view_employee()
    elif choice == '4':
        update_employee()
    elif choice == '5':
        delete_employee()
    elif choice == '0':
        exit()
    else:
        print('Invalid choice')
        main_menu()


def add_employee():
    name = input('Enter the employee name: ')
    age = int(input('Enter the employee age: '))
    address = input('Enter the employee address: ')
    job_title = input('Enter the employee job title: ')

  
    employee = {'name': name, 'age': age, 'address': address, 'job_title': job_title}
    result = collection.insert_one(employee)

    print('Employee added with ID:', result.inserted_id)
    main_menu()


def list_employees():
    cursor = collection.find()

    print('Employee List')
    print('=============')
    for employee in cursor:
        print(employee['name'], employee['age'], employee['address'], employee['job_title'])

    main_menu()


from pymongo import MongoClient
from bson import ObjectId

client = MongoClient()
collection = client.mydb.employees

def view_employee():
    employee_id = input('Enter the employee ID: ')
    employee = collection.find_one({'_id': ObjectId(employee_id)})

    if employee:
        print('Employee Details')
        print('===============')
        print('Name:', employee['name'])
        print('Age:', employee['age'])
        print('Address:', employee['address'])
        print('Job Title:', employee['job_title'])
    else:
        print('Employee not found')

    main_menu()
    


def update_employee():
    employee_id = input('Enter the employee ID: ')
    employee = collection.find_one({'_id': ObjectId(employee_id)})

    if employee:
        name = input('Enter the employee name [' + employee['name'] + ']: ') or employee['name']
        age = input('Enter the employee age [' + str(employee['age']) + ']: ') or employee['age']
        address = input('Enter the employee address [' + employee['address'] + ']: ') or employee['address']
        job_title = input('Enter the employee job title [' + employee['job_title'] + ']: ') or employee['job_title']

        
        result = collection.update_one(
            {'_id': ObjectId(employee_id)},
            {'$set': {'name': name, 'age': int(age), 'address': address, 'job_title': job_title}}
        )

        print(result.modified_count, 'employee updated')
    else:
        print('Employee not found')

    main_menu()
def delete_employee():
    employee_id = input('Enter the employee ID: ')
    employee = collection.find_one({'_id': ObjectId(employee_id)})

    if employee:
        result = collection.delete_one({'_id': ObjectId(employee_id)})
        print(result.deleted_count, 'record(s) deleted')
    else:
        print('Employee not found')

    main_menu()
    
main_menu()
```
